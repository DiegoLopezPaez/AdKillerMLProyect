```python
#libreria para trabajar con audio
import librosa, librosa.display
#libreria para graficos
import matplotlib.pyplot as plt
```

```python
file = "rock.00068.wav"
```

```python
#con esto obtenemos los samples para el audio
#sr = sample rate
#signal sera un array con sr * T elementos, donde sr es el valor que
#pasamos por sr, y T es la duracion del archivo de audio que hayamos
#pasado (en segundos). Cada uno de los valores es la amplitud del
#wave form
signal, sr = librosa.load(file, sr=22050)
```

```python
#Con esto podemos ver un grafico de nuestro audio
librosa.display.waveshow(signal, sr=sr)
#Aca marcamos los datos que mostramos, basicamente el grafico muestra
#la amplitud en funcion del tiempo
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
```

Aplicamos la FFT -> FAST FOURIER TRANSFORM

```python
import numpy as np
```

```python
fft = np.fft.fft(signal)

magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(frequency)/2)]

plt.plot(left_frequency, left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

Hasta ahora no tenemos informacion del tiempo. Para obtener informacion del tiempo, lo que tenemos que hacer es un STFT -> SHORT TIME FOURIER TRANSFORM. Es decir, hacer FFT en distintos segmentos del audio a lo largo de todo el audio, asi obtenemos informacion de FFT para distintos momentos en el tiempo del audio.
Nos da informacion de la amplitud en funcion de la frecuencia y tiempo

```python
### numero de samples por cada fft
n_ftt = 2048
#cantidad que estamos moviendonos en el sonido hacia la derecha
#para samplear el siguiente fft
hop_length = 512
stft = librosa.core.stft(signal, hop_length=hop_length, n_fft=n_ftt)
spectogram = np.abs(stft)

log_spectogram = librosa.amplitude_to_db(spectogram)

librosa.display.specshow(log_spectogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
```

```python
mffcs = librosa.feature.mfcc(y=signal, n_mfcc=13, n_fft=n_ftt, hop_length=hop_length)
librosa.display.specshow(mffcs, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("MFCC")
plt.colorbar()
```
